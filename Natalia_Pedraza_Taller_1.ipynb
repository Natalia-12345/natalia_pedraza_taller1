### **1. Visualización de valor agregado**

La primera visualización es a cerca del valor agregado por municipio en miles de millones de pesos a precio corriente de Colombia en el año 2017, representado con barras 3D. La visualización fue obtenida de la pagina de mapas interactivos del DANE [en este link](https://dane.maps.arcgis.com/apps/MapSeries/index.html?appid=1218a78169a14e77b9645f876e1014cc).


#### Abstraer ¿qué?, ¿por qué? y ¿cómo? 

1. ¿Qué?

El tipo de datos es dato de posición.
El tipo de dataset es geométrico.
La disponibilidad del dataset es estática dado a que esta en el año 2017 especificamente.
En tipos de atributos tenemos que los departamentos en los que se divide el mapa son atributos categóricos y el valor agregado de cada uno de los departamentos es un atributo ordenado cuantitativo.

2. ¿Por qué?

Los verbos o acciones son consumir, ya sea descubrir o presentar, e identificar.
En sustantivos u objetivos tenemos que se puede ver la similitud de los atibutos.

3. ¿Cómo? 

En el como se utiliza la codificación de alinear las barras para una mejor comparación.

#### Marcas y canales utilizados

Se utiliza una marca de linea.
Para los canales se utilizan: Canal de volumen tamaño 3D y tambien el canal de saturación del color.

#### Cumple o no cumple alguna de las reglas generales descritas

Para esta visualización no se cumple la regla general del 3D injustificado con la oclusión. Hay otras maneras de hacer cumplir el objetivo de la visualización sin el uso de 3D.

#### Mejoras posibles que le realizaría a la visualización

Con la base del mapa del país seria mejor cambiar la marca de linea por una de puntos. En canal se podria cambiar el volumen 3d por area 2d,dado que la altura de las barras del 3D no son congruentes con los datos que pretenden representar, una barra de 212.289 de valor agregado (Bogotá) tiene la misma altura y saturación de color que una barra de 7884 de valor agregado (Ibagué, Tolima), por lo tanto se descarta que el 3D sea recomendado. 

Adicionalmente en el canal de saturación de color añadir mas tonos a la escala de colores en la saturacion ya que se puede ver un rango muy pequeño que no permite una buena diferenciación si se tiene en cuenta los valores de valor agrgado de cada municipio. 

### **2. Visualización defunciones por COVID-19**

La segunda visualización es es parte de un informe que muestra las defunciones de COVID-19, dividiendolas en casos confirmados , sospechosos y las defunciones por neumonia e influeza desde la semana 10 del 2019 hasta las semana 2 del 2021. [En este link](https://www.dane.gov.co/index.php/estadisticas-por-tema/demografia-y-poblacion/informe-de-seguimiento-defunciones-por-covid-19)


#### Abstraer ¿qué?, ¿por qué? y ¿cómo? 

1. ¿Qué?

Los datos son tres grupos de defunciones por COVID-19 divido en tres grupo: los casos cofirmados, los casos sospechoso y las muertes por neumonia e influenza. Los tres grupos tienen las mismas caracteristicas que se exponen a continuación.

El tipo de datos son atributos.
El tipo de dataset es temporal.
La disponibilidad del dataset es estática dado a que se centra úniamente en el rango e tiempo descrito.
En tipos de atributos los datos son ordenados cuantitativos dado a que son un conteo de casos.

2. ¿Por qué?

Los verbos o acciones son en consumir: presentar, y en consulta puede ser para comprar los diferentes grupos de datos.
En sustantivos u objetivos tenemos que para todos los datos podemos ver la tendencia.

3. ¿Cómo? 

En el como se utiliza la faceta de sobreponer los tres rupos de datos explicados anteriormente en una misma visualización.

#### Marcas y canales utilizados

Se utiliza la marca de puntos. Para los canales: El canal de posición tanto horizontal como vertical, el canal de forma y el canal de color para diferenciar los grupos de datos.

#### Cumple o no cumple alguna de las reglas generales descritas

La visualización cumple con las reglas generales descritas en clase.

#### Mejoras posibles que le realizaría a la visualización

Para esta visualización considero que seria bueno cambiar el canal de color para el grupo de datos de muertes por neumonia o influenza ya que este color se puede llegar a confundir con los ejes del planoy es dificil de distinguir comparado con los otros tonos.
